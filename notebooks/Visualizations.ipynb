```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
# Separate visualizations for each metric
metrics = ['MPJSE', 'PCK', 'Retrieval_Time']

for metric in metrics:
    plt.figure(figsize=(10, 6))
    for distance_metric in knn_results['Distance_Metric'].unique():
        subset = knn_results[knn_results['Distance_Metric'] == distance_metric]
        plt.plot(subset['K_value'], subset[metric], label=distance_metric, marker='o')

    plt.title(f'{metric} across Different K-values and Distance Metrics')
    plt.xlabel('K_value')
    plt.ylabel(metric)
    plt.legend()
    plt.show()

# Statistical Analysis using ANOVA
for metric in metrics:
    anova_result = f_oneway(
        *[knn_results[knn_results['Distance_Metric'] == distance_metric][metric] for distance_metric in knn_results['Distance_Metric'].unique()]
    )
    print(f"ANOVA for {metric}:")
    print("F-statistic:", anova_result.statistic)
    print("P-value:", anova_result.pvalue)
    print("\n")
```

```python
import pandas as pd

# Assuming your data is stored in a DataFrame called 'df'
# Replace this with the actual DataFrame you have

# Identify the best and worst values for each metric
best_values = knn_results.groupby(['Distance_Metric']).apply(lambda group: group.loc[group['MPJSE'].idxmin()]).reset_index(drop=True)
worst_values = knn_results.groupby(['Distance_Metric']).apply(lambda group: group.loc[group['MPJSE'].idxmax()]).reset_index(drop=True)

# Create a summary table for the best values
best_table = pd.DataFrame({
    'Distance_Metric': best_values['Distance_Metric'],
    'Best_K_value': best_values['K_value'],
    'Best_MPJSE': best_values['MPJSE'],
    'Best_PCK': best_values['PCK'],
    'Best_Retrieval_Time': best_values['Retrieval_Time']
})

# Create a summary table for the worst values
worst_table = pd.DataFrame({
    'Distance_Metric': worst_values['Distance_Metric'],
    'Worst_K_value': worst_values['K_value'],
    'Worst_MPJSE': worst_values['MPJSE'],
    'Worst_PCK': worst_values['PCK'],
    'Worst_Retrieval_Time': worst_values['Retrieval_Time']
})

# Display the tables
print("Best Values:")
print(best_table)

print("\nWorst Values:")
print(worst_table)
```
